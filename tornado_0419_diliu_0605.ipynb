```python
import numpy as np
import sympy as sp
from scipy.optimize import minimize_scalar
#  y = ax^2 + bx + c
def quadratic_function(x, a, b, c):
    return a*x**2 + b*x + c

# 计算垂线与二次函数的交点
def intersection_point(a, b, c, x0, y0):
    x = sp.symbols('x')
    y = quadratic_function(x, a, b, c)
    # 方程为 (x - x0) * y'(x0) + (y - y0) = 0
    dydx = sp.diff(y, x)
    line_eq = (x - x0) * dydx.subs(x, x0) + (y - y0)
    intersection_x = sp.solve(line_eq, x)
    intersection_points = [(float(ix), quadratic_function(float(ix), a, b, c)) for ix in intersection_x]
    return intersection_points

# 计算两个点之间的距离
def distance(point1, point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# 目标函数：计算直线上点 (x0, y0) 和对应的二次函数交点之间的距离
def objective(x0, a, b, c, m, c_line):
    y0 = m * x0 + c_line
    intersection_points = intersection_point(a, b, c, x0, y0)
    distances = [distance((x0, y0), p) for p in intersection_points]
    return -max(distances)  # 返回负值以便使用最小化方法

# 在区间 [x1, x2] 上找到最大距离
def find_max_distance(a, b, c, m, c_line, x1, x2):
    result = minimize_scalar(objective, bounds=(x1, x2), args=(a, b, c, m, c_line), method='bounded')
    x0_max = result.x
    y0_max = m * x0_max + c_line
    intersection_points = intersection_point(a, b, c, x0_max, y0_max)
    distances = [distance((x0_max, y0_max), p) for p in intersection_points]
    max_distance = max(distances)
    max_distance_point = intersection_points[np.argmax(distances)]
    return max_distance, (x0_max, y0_max), max_distance_point

# 示例参数
a, b, c = 0.42830988029573763, 84.24173532596292, 4176.647152764763  # 二次函数 y = x^2 - 2x + 1
m, c_line = d, 0   # 直线 y = x
x1, x2 = -1, 3     # 区间

max_distance, point_on_line, intersection_point = find_max_distance(a, b, c, m, c_line, x1, x2)
print("最大距离:", max_distance)
print("直线上的点坐标:", point_on_line)
print("二次函数上的交点坐标:", intersection_point)
```
